```python
import plotly.express as px
import numpy as np
```

```python
N = 256
h = 700
I, J= np.indices((N, N))
fig = px.imshow(np.mod(I|J, 19), color_continuous_scale='deep_r') 
fig.update_layout(width=h, height=h, title_text="$\\text{Heatmap for }\: a_{ij}=\\text{i | j % 19}$", 
                  title_x=0.5, xaxis_visible=False, yaxis_visible=False,
                  coloraxis_showscale=False, font_family="Open-Sherif");

fig.show()
```
